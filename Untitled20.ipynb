```python
noisy = gray + np.random.randn(*gray.shape) * 25
```

```python
import matplotlib.pyplot as plt

plt.imshow(noisy, cmap='gray')
plt.title("Noisy Image")
plt.axis('off')
plt.show()
```

```python


avg_blur = cv2.blur(noisy, (7 ,7))


plt.imshow(avg_blur, cmap='gray')
plt.title("Average Blur")
plt.axis('off')
plt.show()
```

```python
gaussian_blur = cv2.GaussianBlur(noisy, (7, 7), 0)


plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
noisy_uint8 = noisy.astype('uint8')

median_blur = cv2.medianBlur(noisy_uint8, 7)

plt.imshow(median_blur, cmap='gray')
plt.title("Median Blur")
plt.axis('off')
plt.show()
```

Gaussian blur handled the noise better

Gaussian blur handled the edges best

```python
h, w = img.shape[:2]


resized = cv2.resize(img, (w//2, h//2))


center = (w//2, h//2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, rotation_matrix, (w, h))
translation_matrix = [[1, 0, 50],
                      [0, 1, 30]]
translated = cv2.warpAffine(img,
                             np.float32(translation_matrix),
                             (w, h))
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.title("Original")
plt.imshow(img, cmap='gray' if len(img.shape)==2 else None)
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Resized (Half)")
plt.imshow(resized, cmap='gray' if len(resized.shape)==2 else None)
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Rotated 45")
plt.imshow(rotated, cmap='gray' if len(rotated.shape)==2 else None)
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Translated (50,30)")
plt.imshow(translated, cmap='gray' if len(translated.shape)==2 else None)
plt.axis('off')

plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


_, global_thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)


adaptive_thresh = cv2.adaptiveThreshold(
    gray,
    255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)


_, otsu_thresh = cv2.threshold(
    gray,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.title("Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1,4,2)
plt.title("Global Threshold")
plt.imshow(global_thresh, cmap='gray')
plt.axis('off')

plt.subplot(1,4,3)
plt.title("Adaptive Threshold")
plt.imshow(adaptive_thresh, cmap='gray')
plt.axis('off')

plt.subplot(1,4,4)
plt.title("Otsu Threshold")
plt.imshow(otsu_thresh, cmap='gray')
plt.axis('off')

plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


_, binary = cv2.threshold(
    gray, 0, 255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)
binary = cv2.bitwise_not(binary)


masked = cv2.bitwise_and(img, img, mask=binary)


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
masked_rgb = cv2.cvtColor(masked, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Binary Image")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Masked Output")
plt.imshow(masked_rgb)
plt.axis('off')

plt.show()
```

Real images usually have noise and tiny intensity fluctuations. Smoothing helps clean these up so thresholding doesnt mistake noise for actual objects, giving a cleaner and more reliable result.

HSV separates color from brightness, which makes it easier to isolate objects based on color even when lighting changes. In RGB, color and brightness are mixed together, so segmentation can break under uneven lighting

Yes. Histogram equalization can exaggerate noise and alter brightness patterns, which can confuse thresholding and lead to incorrect segmentation of the object and background

Built-in OpenCV functions are much faster and more reliable because they are optimized at a low level. Writing pixel-by-pixel loops in Python is slower and more likely to cause errors
