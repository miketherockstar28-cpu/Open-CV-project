````python
def blur_gray(Gray_img):

    kernel = np.ones((3, 3)) / 9


    H, W = Gray_img.shape


    padded = np.pad(Gray_img, pad_width=1, mode='constant', constant_values=0)

    print("Padded Image:")
    print(padded)
    print()


    new_img = np.zeros((H, W))


    for i in range(H):
        for j in range(W):

            total = 0.0


            for m in range(3):
                for n in range(3):
                    total += padded[i + m, j + n] * kernel[m, n]

            new_img[i, j] = total

    return new_img

blurred = blur_gray(Gray)
plt.imshow(blurred, cmap='gray')
plt.axis('off')
plt.show()
print("Min, Max:", blurred.min(), blurred.max())
````

A convoulation kernal is a filter it is slid over a image to change intensity value of each pixel.

Changing the kernal changes the pixel value at each pixel which affects the intensity and chnges the output

Because human eyes don't percieve colours equally so to preserve brighness we use weighted averages
